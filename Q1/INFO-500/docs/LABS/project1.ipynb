```python
import matplotlib.pyplot as plt
from skimage.transform import resize
 
def imgWatermark(img_src,wm_src,dest,loc, scale=2, alpha=0.2): 
    fig=plt.figure(figsize=(15,15))
    # READ THE WATERMARK
    watermark = plt.imread(wm_src)
    #Invert the watermark
    for y in range(watermark.shape[0]):
        for x in range(watermark.shape[1]):
            for i in range(watermark.shape[2]-1):
                if (watermark[y,x,i]==0):
                    watermark[y,x,i]=1
                elif(watermark[y,x,i]==1):
                    watermark[y,x,i]=0             
    # RESIZING
    new_size = [int(watermark.shape[0]/scale), int(watermark.shape[1]/scale), watermark.shape[2]]
    watermark = resize(watermark, new_size, mode='constant')
    # TRANSPARENCY
    for y in range(watermark.shape[0]):
        for x in range(watermark.shape[1]):
            if (watermark[y,x,0]==0):
                    watermark[y, x,-1] *= 0
    watermark[:, :,-1] *= alpha
    # READ THE IMAGE
    plt.imshow(plt.imread(img_src))
    # ADD THE WATERMARK
    plt.figimage(watermark, loc[0], loc[1], zorder=10)
    # HIDE THE AXIS
    plt.axis('off')
    # SAVE THE IMAGE
    plt.savefig(dest,bbox_inches='tight')
    return fig
```

```python
for img in ['etretat.jpg']:
    imgWatermark(img_src=img,wm_src='watermark.png',dest='wm_%s'%img, loc=[40,200]);
```
